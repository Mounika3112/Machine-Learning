```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('bayes.csv')
df
```

```python
n=0
y=0
t=len(df)
for x in df['Class']:
    if x=='N':
        n=n+1
    else:
        y=y+1
```

```python
#TRAINING 
def calculateProb(x,it):
    no=yes=0
    array=np.array(df[x])
    index=np.where(array==it)[0]
    for i in index:
        if df['Class'][i]=='N':
            no=no+1
        else:
            yes=yes+1
    return yes/y,no/n
```

```python
prob_yes=y/t
prob_no=n/t
ls_yes=[]
ls_no=[]
#TESTING
for i in range(len(df.columns)-1):
    value=df.columns[i]
    print(value)
    it=input("enter value: ")
    kyes,kno=calculateProb(value,it)
    ls_yes.append(kyes)
    ls_no.append(kno)
for x in ls_yes:
    prob_yes=prob_yes*x
for x in ls_no:
    prob_no=prob_no*x
# print(prob_yes)
# print(prob_no)
if(prob_yes>prob_no):
    print("Play Tennis=YES")
else:
    print("Play Tennis=NO")
```
